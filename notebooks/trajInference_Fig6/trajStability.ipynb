# Generate Figure

```python
vlm.S_sz.shape
```

```python
plt.rcParams.update({'font.size': 22})
fig = plt.figure(figsize=(12,12))
ax=[]
steps = (20,20)
ax.append(plt.subplot2grid((3, 2), (1, 0)))
ax.append(plt.subplot2grid((3, 2), (1, 1)))
ax.append(plt.subplot2grid((3, 2), (2, 0)))
ax.append(plt.subplot2grid((3, 2), (2, 1)))
gridArrowPlots(vlm,Trans=["sqrt"],embed="UMAP",sim=False,ax=ax[0],quiver_scale=10,plot_baseline=False,legend=False,steps=steps,embed_neigh=5)
gridArrowPlots(vlm,Trans=["sqrt"],embed="UMAP",sim=False,ax=ax[1],quiver_scale=10,plot_baseline=False,legend=False,steps=steps,embed_neigh=17)
gridArrowPlots(vlm,Trans=["sqrt"],embed="UMAP",sim=False,ax=ax[2],quiver_scale=10,plot_baseline=False,legend=False,steps=steps,embed_neigh=30)
gridArrowPlots(vlm,Trans=["sqrt"],embed="UMAP",sim=False,ax=ax[3],quiver_scale=10,plot_baseline=False,legend=False,steps=steps,embed_neigh=50)
fig.tight_layout()
fig.savefig('figure/embed_stability_foreb_umap.png',dpi=600)
```
