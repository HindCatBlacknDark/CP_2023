```python
library(methods)
library(scran)
library(batchelor)
```

```python
names = list('/home/tchari/scanorama/data/pancreas/m20.mtx','/home/tchari/scanorama/data/pancreas/m21.mtx')#,'/home/tchari/scanorama/data/pancreas/m22.mtx') #
data.tables <- list()
for (i in 1:length(names)) {
    data.tables[[i]] <- t(as.matrix(Matrix::readMM(names[[i]]))) #as.matrix(read.table(names[[i]], sep="\t"))
}

print("Done loading")

```

```python
ptm <- proc.time()

#log1p(
Xmnn <- mnnCorrect(
    data.tables[[1]],
    data.tables[[2]]
)
#    data.tables[[3]]
print("Done correcting")

print(proc.time() - ptm)

```
